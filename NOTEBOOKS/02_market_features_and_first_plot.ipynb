# Market Features and First Plot

```python
from pathlib import Path

import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import pandas as pd
```

```python
BASE_DIR = Path.cwd().resolve()
if BASE_DIR.name == "NOTEBOOKS":
    BASE_DIR = BASE_DIR.parent

DATA_DIR = BASE_DIR / "DATA"
FIGURES_DIR = BASE_DIR / "FIGURES"

MARKET_FEATURES_FILE = DATA_DIR / "market_features.csv"
EVENTS_FILE = DATA_DIR / "events.csv"
FIGURE_FILE = FIGURES_DIR / "sp500_with_events.png"
```

```python
market_features = pd.read_csv(MARKET_FEATURES_FILE)
market_features["date"] = pd.to_datetime(market_features["date"])
market_features = market_features.sort_values("date").reset_index(drop=True)

market_features.head()
```

```python
market_features.describe(include="all", datetime_is_numeric=True)
```

```python
events = pd.read_csv(EVENTS_FILE)
events["date"] = pd.to_datetime(events["date"])
events = events.sort_values("date").reset_index(drop=True)

events
```

```python
if FIGURE_FILE.exists():
    image = mpimg.imread(FIGURE_FILE)
    plt.figure(figsize=(12, 6))
    plt.imshow(image)
    plt.axis("off")
    plt.show()
else:
    print(f"Figura nao encontrada: {FIGURE_FILE}")
```
