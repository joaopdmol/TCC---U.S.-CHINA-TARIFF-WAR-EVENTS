# Event Windows Inspection

```python
from pathlib import Path

import pandas as pd
```

```python
BASE_DIR = Path.cwd().resolve()
if BASE_DIR.name == "NOTEBOOKS":
    BASE_DIR = BASE_DIR.parent

DATA_DIR = BASE_DIR / "DATA"
EVENTS_FILE = DATA_DIR / "events.csv"
MARKET_FEATURES_FILE = DATA_DIR / "market_features.csv"
EVENT_WINDOWS_FILE = DATA_DIR / "event_windows_long.csv"
```

```python
events = pd.read_csv(EVENTS_FILE)
events["date"] = pd.to_datetime(events["date"])
events = events.sort_values("date").reset_index(drop=True)

market_features = pd.read_csv(MARKET_FEATURES_FILE)
market_features["date"] = pd.to_datetime(market_features["date"])

event_windows = pd.read_csv(EVENT_WINDOWS_FILE)
for column in ["event_date", "anchor_date", "market_date"]:
    event_windows[column] = pd.to_datetime(event_windows[column])

events.head(), market_features.head(), event_windows.head()
```
