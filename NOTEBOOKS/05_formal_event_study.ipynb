# Formal Event Study

```python
from pathlib import Path

import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import pandas as pd
```

```python
BASE_DIR = Path.cwd().resolve()
if BASE_DIR.name == "NOTEBOOKS":
    BASE_DIR = BASE_DIR.parent

DATA_DIR = BASE_DIR / "DATA"
FIGURES_DIR = BASE_DIR / "FIGURES"

ABNORMAL_RETURNS_FILE = DATA_DIR / "abnormal_returns_long.csv"
FORMAL_CAR_BY_EVENT_FILE = DATA_DIR / "formal_car_by_event.csv"
FORMAL_CAR_BY_GROUP_FILE = DATA_DIR / "formal_car_by_group.csv"
FORMAL_CAR_BY_EVENT_FIGURE = FIGURES_DIR / "formal_car_by_event.png"
FORMAL_CAR_BY_GROUP_FIGURE = FIGURES_DIR / "formal_car_by_group.png"
ABNORMAL_RETURNS_BY_GROUP_FIGURE = FIGURES_DIR / "abnormal_returns_by_group.png"
```

```python
abnormal_returns = pd.read_csv(ABNORMAL_RETURNS_FILE)
formal_car_by_event = pd.read_csv(FORMAL_CAR_BY_EVENT_FILE)
formal_car_by_group = pd.read_csv(FORMAL_CAR_BY_GROUP_FILE)

abnormal_returns.head(), formal_car_by_event.head(), formal_car_by_group.head()
```

```python
formal_car_by_group.describe(include="all")
```

```python
example_event_id = "E10"
abnormal_returns.loc[abnormal_returns["event_id"] == example_event_id].sort_values(["window_label", "relative_day"]).reset_index(drop=True)
```

```python
for figure_path in [FORMAL_CAR_BY_EVENT_FIGURE, FORMAL_CAR_BY_GROUP_FIGURE, ABNORMAL_RETURNS_BY_GROUP_FIGURE]:
    if figure_path.exists():
        image = mpimg.imread(figure_path)
        plt.figure(figsize=(12, 6))
        plt.imshow(image)
        plt.title(figure_path.name)
        plt.axis("off")
        plt.show()
    else:
        print(f"Figura nao encontrada: {figure_path}")
```
