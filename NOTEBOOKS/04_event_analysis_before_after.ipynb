# Event Analysis Before/After

```python
from pathlib import Path

import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import pandas as pd
```

```python
BASE_DIR = Path.cwd().resolve()
if BASE_DIR.name == "NOTEBOOKS":
    BASE_DIR = BASE_DIR.parent

DATA_DIR = BASE_DIR / "DATA"
FIGURES_DIR = BASE_DIR / "FIGURES"

SUMMARY_BY_EVENT_FILE = DATA_DIR / "event_summary_by_event.csv"
SUMMARY_BY_GROUP_FILE = DATA_DIR / "event_summary_by_group.csv"
CAR_BY_EVENT_FILE = DATA_DIR / "car_by_event.csv"
VOLATILITY_BY_EVENT_FILE = DATA_DIR / "volatility_by_event.csv"
CAR_FIGURE_FILE = FIGURES_DIR / "car_by_event.png"
RETURNS_FIGURE_FILE = FIGURES_DIR / "returns_by_group.png"
VOLATILITY_FIGURE_FILE = FIGURES_DIR / "volatility_by_group.png"
```

```python
summary_by_event = pd.read_csv(SUMMARY_BY_EVENT_FILE)
summary_by_group = pd.read_csv(SUMMARY_BY_GROUP_FILE)
car_by_event = pd.read_csv(CAR_BY_EVENT_FILE)
volatility_by_event = pd.read_csv(VOLATILITY_BY_EVENT_FILE)

summary_by_event.head(), summary_by_group.head(), car_by_event.head(), volatility_by_event.head()
```

```python
summary_by_group[[
    "event_group",
    "window_label",
    "period",
    "sp500_return_mean",
    "nasdaq_return_mean",
    "shanghai_return_mean",
    "event_count",
]].sort_values(["event_group", "window_label", "period"]).reset_index(drop=True)
```

```python
example_event_id = "E10"
summary_by_event.loc[summary_by_event["event_id"] == example_event_id].sort_values(["window_label", "period"]).reset_index(drop=True)
```

```python
for figure_path in [CAR_FIGURE_FILE, RETURNS_FIGURE_FILE, VOLATILITY_FIGURE_FILE]:
    if figure_path.exists():
        image = mpimg.imread(figure_path)
        plt.figure(figsize=(12, 6))
        plt.imshow(image)
        plt.title(figure_path.name)
        plt.axis("off")
        plt.show()
    else:
        print(f"Figura nao encontrada: {figure_path}")
```
